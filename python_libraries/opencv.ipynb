```python
import cv2
```

```python
image = cv2.imread("car.jpg")
```

```python
print(type(image))
```

```python
cv2.imshow("image name", image)
```

```python
cv2.waitKey(0)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 

plt.imshow(image)
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(image_gray, cmap="gray")
```

```python
small_im= cv2.resize(image_rgb, (100,100))
```

```python
plt.imshow(small_im)
```

```python
ksize =  (15,25)

#using blur method
img_blur = cv2.blur(image_rgb,ksize)
fig, (ax1, ax2) = plt.subplots(2,1, figsize = (15,10))
ax1.imshow(image_rgb)

ax2.imshow(img_blur)
```

```python
img_edge = cv2.Canny(image_rgb,125,175) #2 threshold
plt.imshow(img_edge)
```

```python
# make rectangle
# represents top left corner of rectangle
start_point = (110, 200)

# represents bottom right corner of rectangle
end_point= (700,550)

# red color in RGB  
color = (255, 0 , 0)
```

```python
line_thickness = 5
```

```python
img_rectangle = cv2.rectangle(image_rgb, start_point, end_point, color, line_thickness)
```

```python
plt.imshow(img_rectangle)
```
